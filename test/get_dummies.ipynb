```python
import pandas as pd
import numpy as np

fruit = pd.DataFrame({
    'name': ['apple', 'banana', 'cherry', 'durian', np.nan],
    'color': ['red', 'yellow', 'red', 'green', np.nan]
})

pd.get_dummies(fruit, dtype='int')
```
