```python
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder

iris = sns.load_dataset('iris')
enc = OrdinalEncoder()
enc.fit(iris[['species']])

iris_onehot = enc.transform(iris[['species']])
print(iris_onehot)
```
