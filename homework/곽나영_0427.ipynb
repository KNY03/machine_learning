```python
# 기본 라이브러리 불러오기
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# 데이터 불러오기
df = pd.read_csv("./titanic/train.csv")

# 원본 데이터 확인
df.head()
```

```python
df[["Age", "Embarked"]].isnull().sum()
```

```python
# Age 결측값을 평균값으로 처리
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Embarked 결측값을 최빈값으로 처리
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# 결측값 처리 결과를 데이터프레임에 저장
missing_result_df = df.copy()

missing_result_df.head()
```

```python
# 결측값 처리 확인
missing_result_df[["Age", "Embarked"]].isnull().sum()
```

```python
# 기존 데이터프레임을 새로운 데이터프레임으로 저장
label_df = df.copy()

# LabelEncoder 객체 생성
sex_encoder = LabelEncoder()
embarked_encoder = LabelEncoder()

# Sex, Embarked 레이블 인코딩
sex_encoded = sex_encoder.fit_transform(label_df["Sex"])
embarked_encoded = embarked_encoder.fit_transform(label_df["Embarked"])

# 기존 Sex, Embarked 컬럼 삭제
label_df = label_df.drop(["Sex", "Embarked"], axis=1)

# 인코딩 결과를 새로운 컬럼으로 추가
label_df["Sex_Label"] = sex_encoded
label_df["Embarked_Label"] = embarked_encoded

# 결과 데이터프레임 저장
label_result_df = label_df.copy()

label_result_df.head()
```

```python
# 기존 데이터프레임을 새로운 데이터프레임으로 저장
onehot_df = df.copy()

# 원핫인코딩할 컬럼 선택
onehot_target = onehot_df[["Sex", "Embarked"]]

# OneHotEncoder 객체 생성
encoder = OneHotEncoder(sparse_output=False)

# 원핫인코딩 수행
onehot_array = encoder.fit_transform(onehot_target)

# 원핫인코딩된 컬럼명 생성
onehot_columns = encoder.get_feature_names_out(["Sex", "Embarked"])

# 원핫인코딩 결과를 데이터프레임으로 변환
onehot_result = pd.DataFrame(onehot_array, columns=onehot_columns)

# 기존 Sex, Embarked 컬럼 삭제
onehot_df = onehot_df.drop(["Sex", "Embarked"], axis=1)

# 인덱스 맞추기
onehot_result.index = onehot_df.index

# 원핫인코딩 결과를 기존 데이터프레임에 추가
onehot_df = pd.concat([onehot_df, onehot_result], axis=1)

# 결과 데이터프레임 저장
onehot_result_df = onehot_df.copy()

onehot_result_df.head()
```

```python
print(missing_result_df )  # 결측값 처리 결과
print(label_result_df)     # 레이블 인코딩 결과
print(onehot_result_df)    # 원핫인코딩 결과
```
